## Income and Expense Relationships:

    1-Is there a correlation between monthly income and annual income?
    2-Is there a correlation between the number of earning members and monthly household income?
    3-How does the monthly household expense relate to the number of family members?

```python
mnthly_inc = income['Mthly_HH_Income']
```

```python
annual_inc = income['Annual_HH_Income']
```

```python
np.corrcoef(mnthly_inc,annual_inc)[0,1]
```

- There is a `strong +ve` correlation,because the value is colse to 1

```python
no_of_earn_mem = income['No_of_Earning_Members']
```

```python
np.corrcoef(no_of_earn_mem,mnthly_inc)[0,1]
```

- There is a `+ve` correlation,Because the value is +ve

```python
mnthly_exp = income['Mthly_HH_Expense']
```

```python
no_of_fam_mem = income['No_of_Fly_Members']
```

```python
income[['Mthly_HH_Expense','No_of_Fly_Members']].corr()
```

```python
np.corrcoef(mnthly_exp,no_of_fam_mem)[0,1]
```

- There is a +ve relation
